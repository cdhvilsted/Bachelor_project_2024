```python
import mne
import os
import matplotlib.pyplot as plt
import sklearn
from examples.split_based_on_resting import LoadFile
from mne_icalabel import label_components
import numpy as np
```

Reading data as .edf file (not preprocessed, run cell with pyPREP etc.)

```python
# Importing data
path ="EDF_test/"
folder = os.fsencode(path)
patient_number = 2
raw_liste = []

#finding all files for specific person
for file in os.listdir(folder):
    filename = os.fsdecode(file)
    print(filename)
    fname = path+filename
    person_number = filename[:2]
    if person_number[0] == 0:
        person_number = person_number[1]
    pn = int(person_number)
    if pn == patient_number:
        raw, annot, annot_dict = LoadFile(fname)
        raw_liste.append(raw)
    if pn > patient_number:
        break
#checking if multiple files should be concatenated
if len(raw_liste) != 1:
    raw_con = mne.concatenate_raws(raw_liste)
else:
    raw_con = raw_liste[0]
```

```python
import pyprep

prep_params = {
    'ref_chs': 'eeg',
    'reref_chs': 'eeg',
    'line_freqs': [50]
}

prep = pyprep.PrepPipeline(raw_con, prep_params, montage='standard_1020', random_state=97)
prep.fit()

print('Bad channels: ', format(prep.interpolated_channels))
```

```python
# Setting bad channels
#raw_con.info['bads'].extend(['P7', 'F7'])
```

```python
raw_con = prep.raw
```

```python
raw.ch_names
#=['Fp1', 'Fp2', 'F3', 'F4', 'C3', 'C4', 'P3', 'P4', 'O1', 'O2', 'F7', 'F8', 'T7', 'T8', 'P7', 'P8', 'Fz', 'Cz', 'Pz', 'ECG EKG']
```

```python
# Interpolating bad channels
#raw = raw.copy().interpolate_bads(reset_bads=True, verbose=True, mode='accurate')


# Band-pass filtering
raw_con = raw_con.copy().filter(l_freq=0.1, h_freq=45, verbose=False)
# Down-sampling
raw_con = raw_con.copy().resample(sfreq = 250)
```

```python
# Plot after bad channel rejection, band-pass filtering and downsampling
raw_con.plot()
plt.show()
```

```python
# ICA
# Finding all components (19 channels)
ica = mne.preprocessing.ICA(n_components=18, random_state=97, max_iter=800)
ica.fit(raw_con)
```

```python
# Labelling ICA components
component_dict = label_components(raw_con, ica, method='iclabel')
component_dict
```

```python
import collections

c = collections.Counter(component_dict['labels'])
print(c)
```

```python
prob = list(component_dict['y_pred_proba'])
labels = list(component_dict['labels'])

bad_comp = []

for i in range(len(labels)):
    if labels[i] != 'brain':
        if prob[i] >= 0.5:
            bad_comp.append(i)

ica.exclude.extend(bad_comp)
ica.apply(raw_con)
```

```python
# Plotting scalp maps
ica.plot_components(title='ICA components for patient 2')
plt.show()
```
